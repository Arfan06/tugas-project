```python
#Untuk menggabungkan dataframe item dan order
order_customers_df = pd.merge(
    left=items_df,
    right=order_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
order_customers_df.head()
```

```python
#Untuk menggabungkan dataframe order+item dengan dataframe customers
all_key_df = pd.merge(
    left=order_customers_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
all_key_df.head()
```

```python
#Untuk menampilkan kota dengan transaksi
all_key_df.groupby(by= 'customer_city').order_id.nunique().sort_values(ascending=False)
```

```python
#Untuk menggabungkan dataframe order+item+customers dengan dataframe review
akhir_df = pd.merge(
    left=all_key_df,
    right=reviews_df,
    how='left',
    left_on='order_id',
    right_on='order_id'
)

akhir_df.head()
```

```python
#Untuk menampilkan total score
akhir_df.groupby(by= 'review_score').product_id.nunique().sort_values(ascending=False)
```

```python
#Untuk menggabungkan dataframe product dengan kategori
product_category_df = pd.merge(
    left=product_df,
    right=category_df,
    how='left',
    left_on='product_category_name',
    right_on='product_category_name'
)

product_category_df.head()
```

```python
#Untuk menggabungkan semua dataframe
all_df = pd.merge(
    left=akhir_df,
    right=product_category_df,
    how='left',
    left_on='product_id',
    right_on='product_id'
)

all_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Revenue yang Diperoleh Setiap Bulannya?

```python
#melihat pergerakan grafik perbulan
monthly_orders_df = all_df.resample(rule='M', on='order_purchase_timestamp').agg({
    "order_id": "nunique",
    "price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)

plt.figure(figsize=(25, 5))
plt.plot(monthly_orders_df["order_purchase_timestamp"], monthly_orders_df["order_count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Revenue of Orders per Month", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 2: Apa 10 Produk dengan Performa Penjualan Terbaik dan Terburuk?

```python
sum_items_df = all_df.groupby("product_category_name").product_id.nunique().sort_values(ascending=False).reset_index()
sum_items_df.head(21)
```

```python
#Untuk menampilkan kanvas
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

#Untuk setting warna
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

#Untuk membuat gambar pertama
sns.barplot(x="product_id", y="product_category_name", data=sum_items_df.head(10), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Performa Produk Terbaik", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

#Untuk membuat gambar kedua
sns.barplot(x="product_id", y="product_category_name", data=sum_items_df.sort_values(by="product_id", ascending=True).head(10), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Performa Produk Terburuk", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Perbandingan Performa Penjualan", fontsize=20)
plt.show()
```
